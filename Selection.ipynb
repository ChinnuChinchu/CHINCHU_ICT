```python
# Drop unnecessary columns
data.drop(['player_id','behaviour_rating'],axis=1,inplace=True)
```

```python
#Handling missing values
data['gender'].fillna(data['gender'].mode()[0], inplace=True)
data['weight'].fillna(data['weight'].mode()[0], inplace=True)
data['ball_controlling_skills'].fillna(data['ball_controlling_skills'].mode()[0], inplace=True)
data['jumping_skills'].fillna(data['jumping_skills'].mode()[0], inplace=True)
data['penalties_conversion_rate'].fillna(data['penalties_conversion_rate'].mode()[0], inplace=True)
data['mental_strength'].fillna(data['mental_strength'].mode()[0], inplace=True)
data['shot_accuracy'].fillna(data['shot_accuracy'].mode()[0], inplace=True)
data['strong_foot'].fillna(data['strong_foot'].mode()[0], inplace=True)
data['matches_played'].fillna(data['matches_played'].mode()[0], inplace=True)
data['fitness_rating'].fillna(data['fitness_rating'].mode()[0], inplace=True)
data['coaching'].fillna(data['coaching'].mode()[0], inplace=True)
data['years_of_experience'].fillna(data['years_of_experience'].mode()[0], inplace=True)
```

```python
# Encoding

data['country'].value_counts().head(10)
```

```python
other_country = data['country'].value_counts().to_dict()
other_count=[]
for key,val in other_country.items():
    if val<=500:
        other_count.append(key)
```

```python
data.info()
```

```python
data['country'] = data['country'].replace(other_count,'Other')
data['height'] = data['height'].str.replace("'", ".").astype(float)
data.shape
```
